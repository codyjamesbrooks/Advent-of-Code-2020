```python
filename = 'puzzle_eight.txt'
with open(filename) as f:
    data = f.readlines()
for counter in range(len(data)):
    data[counter] = data[counter].rstrip('\n')
    
data
```

```python
def pos_or_neg(num_string):
    if num_string[0] == '+':
        return 1
    if num_string[0] == '-':
        return -1

accumlator = 0
counter = 0
list_of_prev_counter = []

while counter not in list_of_prev_counter:
    list_of_prev_counter.append(counter)
    action = data[counter]
    
    type_of_action = action.split(' ')[0]
    number = pos_or_neg(action.split(' ')[1][0]) * int(action.split(' ')[1][1:])
    
    if type_of_action == 'acc':
        accumlator += number
        counter +=1
        continue
    
    if type_of_action == 'jmp':
        counter += number
        continue
    
    if type_of_action == 'nop':
        counter += 1
        
print(accumlator)
```

```python
def run_program(counter, list_of_prev_counters, accumlator, data=data):
    while counter not in list_of_prev_counter:
        list_of_prev_counter.append(counter)
        action = data[counter]

        type_of_action = action.split(' ')[0]
        number = pos_or_neg(action.split(' ')[1][0]) * int(action.split(' ')[1][1:])

        if type_of_action == 'acc':
            accumlator += number
            counter +=1

        if type_of_action == 'jmp':
            counter += number

        if type_of_action == 'nop':
            counter += 1
        
        if counter == 646:
            print('You have sucessfully finished the game.')
            print(f'The value of the accumlator is {accumlator}')
            return True
```

```python
# Goal is to get the counter to be 646
# i.e. change one jmp or nop to get the counter to be 646

counter = 0
list_of_counters = []
list_of_prev_counters = []
list_of_change_counters = []


while counter not in list_of_prev_counters:
    list_of_prev_counters.append(counter)
    action = data[counter]

    type_of_action = action.split(' ')[0]
    number = pos_or_neg(action.split(' ')[1][0]) * int(action.split(' ')[1][1:])

    if type_of_action == 'acc':
        counter += 1
        accumlator += number
        continue

    if type_of_action == 'jmp' and counter not in list_of_change_counters:
        print(f'Attempt Change of {data[counter]}')
        list_of_change_counters.append(counter)
        counter += 1
        Flag = run_program(counter, list_of_prev_counters, accumlator)
        if Flag == None:
            print('Failed Change. Restarting')
            counter = 0
            list_of_prev_counters = []
            accumlator = 0
            continue

        if Flag == True:
            break
            
    if type_of_action == 'jmp' and counter in list_of_change_counters:
        counter += number
            
    if type_of_action == 'nop' and counter not in list_of_change_counters:
        print(f'Attempt Change of {data[counter]}')
        list_of_change_counters.append(counter)
        counter += number
        Flag = run_program(counter, list_of_prev_counters, accumlator)
        if Flag == None:
            print('Failed Change. Restarting')
            counter = 0
            list_of_prev_counters = []
            accumlator = 0
            continue
        
        if Flag == True:
            break
    
    if type_of_action == 'nop' and counter in list_of_change_counters:
        counter += 1
```
