```python
filename = 'puzzle_eleven.txt'
with open(filename) as f:
    data = f.readlines()

for counter in range(len(data)):
    data[counter] = data[counter].rstrip()
```

```python
def update_seat(x, y, data=data):
    num_occupy = 0
    num_empty = 0
    
    seat_status = data[y][x]
    adj_seat_status = []
    
    if (0 < x < (len(data[0]) - 1)) and  (0 < y < (len(data) - 1)):
        slice_above = data[y-1][x-1:x+2]
        slice_below = data[y+1][x-1:x+2]
        sides = data[y][x-1:x+2:2]
    
    if y == 0:
        if x == 0:
            slice_above = []
            slice_below = data[y+1][x:x+2]
            sides = data[y][x+1]
        elif 0 < x < (len(data[0]) - 1):
            slice_above = []
            slice_below = data[y+1][x-1:x+2]
            sides = data[y][x-1:x+2:2]
        else:
            slice_above = []
            slice_below = data[y+1][x-1:]
            sides = data[y][x-1]
            
    if y == len(data) - 1:
        if x == 0:
            slice_above = data[y-1][x:x+2]
            slice_below = []
            sides = data[y][x+1]
        elif 0 < x < (len(data[0]) - 1):
            slice_above = data[y-1][x-1:x+2]
            slice_below = []
            sides = data[y][x-1:x+2:2]
        else:
            slice_above = data[y-1][x-1:]
            slice_below = []
            sides = data[y][x-1]
            
    if 0 < y < len(data) - 1 and x == 0:
        slice_above = data[y-1][x:x+2]
        slice_below = data[y+1][x:x+2]
        sides = data[y][x+1]
    if 0 < y < len(data) -1 and x == len(data[0]) - 1:
        slice_above = data[y-1][x-1:]
        slice_below = data[y+1][x-1:]
        sides = data[y][x-1]
            
    adj_seat_status.append(slice_above)
    adj_seat_status.append(sides)
    adj_seat_status.append(slice_below)
    
    for row in adj_seat_status:
        for seat in row:
            if seat == 'L':
                num_empty += 1
            if seat == '#':
                num_occupy += 1
        
    if seat_status == 'L' and num_occupy == 0:
        return '#'
    if seat_status == '#' and num_occupy >= 4:
        return 'L'
    else:
        return seat_status
```

```python
filename = 'puzzle_eleven.txt'
with open(filename) as f:
    data = f.readlines()

for counter in range(len(data)):
    data[counter] = data[counter].rstrip()

flag = True
counter = 0

while flag:
    updated_seating = []
    for y in range(len(data)):
        new_row = []
        for x in range(len(data[0])):
            new_row.append(update_seat(x,y,data))
        updated_seating.append(new_row)

    counter += 1
    print(f'\nThe seating arrangment after {counter} rounds.')
    num_occupied = 0
    for row in updated_seating:
        for seat in row:
            if seat == '#':
                num_occupied += 1
    print(f'The number of occupied seat is {num_occupied}.')
    
    if updated_seating == data:
        print('The seats have stopped changing.')
        print(f'The final number of occupied seats is {num_occupied}.')
        flag = False
    
    data = updated_seating
```

```python
def update_seat_2(x, y, data=data):
    num_occupy = 0
    
    seat_status = data[y][x]
    line_of_sight_seats = []
    look_dir = 1
    
    while x + look_dir <= len(data[0])-1:
        target = data[y][x+look_dir]
        if target == '#' or target == 'L':
            line_of_sight_seats.append(target)
            look_dir = 1
            break
        elif target == '.' and x + look_dir == len(data[0]) - 1:
            look_dir = 1
            break
        else:
            look_dir += 1
    look_dir = 1
    while x + look_dir <= len(data[0])-1 and y + look_dir <= len(data)-1:
        target = data[y+look_dir][x+look_dir]
        if target == '#' or target == 'L':
            line_of_sight_seats.append(target)
            look_dir = 1
            break
        elif target == '.' and (x + look_dir == len(data[0]) - 1 or y + look_dir == len(data) -1):
            look_dir = 1
            break
        else:
            look_dir += 1
    look_dir = 1
    while y + look_dir <= len(data) - 1:
        target = data[y+look_dir][x]
        if target == '#' or target == 'L':
            line_of_sight_seats.append(target)
            look_dir = 1
            break
        elif target == '.' and y + look_dir == len(data) - 1:
            look_dir = 1
            break
        else:
            look_dir += 1
    look_dir = 1
    while x - look_dir >= 0 and y + look_dir <= len(data) - 1:
        target = data[y+look_dir][x-look_dir]
        if target == '#' or target == 'L':
            line_of_sight_seats.append(target)
            look_dir = 1
            break
        elif target == '.' and ((x - look_dir == len(data[0]) - 1) or (y + look_dir == len(data) -1)):
            look_dir = 1
            break
        else:
            look_dir += 1
    look_dir = 1
    while x - look_dir >= 0:
        target = data[y][x-look_dir]
        if target == '#' or target == 'L':
            line_of_sight_seats.append(target)
            look_dir = 1
            break
        elif target == '.' and x - look_dir == 0:
            look_dir = 1
            break
        else:
            look_dir += 1
    look_dir = 1
    while x - look_dir >= 0 and y - look_dir >= 0:
        target = data[y-look_dir][x-look_dir]
        if target == '#' or target == 'L':
            line_of_sight_seats.append(target)
            look_dir = 1
            break
        elif target == '.' and (x - look_dir == 0 or y - look_dir == 0):
            look_dir = 1
            break
        else:
            look_dir += 1
    look_dir = 1            
    while y - look_dir >= 0:
        target = data[y-look_dir][x]
        if target == '#' or target == 'L':
            line_of_sight_seats.append(target)
            look_dir = 1
            break
        elif target == '.' and y - look_dir == 0:
            look_dir = 1
            break
        else:
            look_dir += 1
    look_dir = 1            
    while x + look_dir <= len(data[0]) - 1 and y - look_dir >= 0:
        target = data[y-look_dir][x+look_dir]
        if target == '#' or target == 'L':
            line_of_sight_seats.append(target)
            look_dir = 1
            break
        elif target == '.' and (x + look_dir == len(data[0]) - 1 or y - look_dir == 0):
            look_dir = 1
            break
        else:
            look_dir += 1

    for seat in line_of_sight_seats:
        if seat == '#':
            num_occupy += 1
        
    if seat_status == 'L' and num_occupy == 0:
        return '#'
    if seat_status == '#' and num_occupy >= 5:
        return 'L'
    else:
        return seat_status
```
