```python
filename = 'puzzle_seven.txt'
with open(filename) as f:
    data = f.readlines()
for counter in range(len(data)):
    data[counter] = data[counter].rstrip('\n')

rules = {}
remove = [' bags', ' bag', '.']
for rule in data:
    for x, y in zip(rule.split(' bags contain ')[0:1], rule.split(' bags contain ')[1:]):
        rules[x] = y
for key, value in rules.items():
    for word in remove:
        rules[key] = rules[key].replace(word, '')
rules
```

```python
shiny_golds = ['shiny gold']
bags = []


for key in rules:
    bags.append(key)

print(len(bags))
    
length = len(shiny_golds)
prev_length = 0

while length > prev_length:
    for bag in bags:
        if bag in shiny_golds:
            continue
        contents = rules[bag].split(', ')
        for x in contents:
            if x in shiny_golds:
                shiny_golds.append(bag)
    strip_dups = []
    [strip_dups.append(x) for x in shiny_golds if x not in strip_dups]
    shiny_golds = strip_dups
    prev_length = length
    length = len(shiny_golds)

print(len(set(shiny_golds)) - 1)
```

```python
winners = []
for key,value in rules.items():
    contents = value.split(', ')
    for bag in contents:
        if bag == 'shiny gold':
            winners.append(key)

print(len(winners))
            
for key,value in rules.items():
    contents = value.split(', ')
    for bag in contents:
        if bag in winners:
            rules[key] = rules[key].replace(bag, 'shiny gold')
```

```python
def number_of_bag_contents(bag, count=0, rules=rules):
    contents = rules[bag].split(', ')
    print(contents)
    for x in contents:
        if x == 'no other':
            return 1
        else:
            print(count, '\n')
            num_bags = 0
            for x in contents:
                num_bags += int(x[0])            
            count += int(x[0]) * number_of_bag_contents(x[2:], count)
    
    return count       
    
```
