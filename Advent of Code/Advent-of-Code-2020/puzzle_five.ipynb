```python
filename = 'puzzle_five.txt'
with open(filename) as f:
    data = f.readlines()
    for counter in range(len(data)): 
        data[counter] = data[counter].rstrip('\n')
```

```python
row_numbers = []
col_numbers = []
seat_ids = []

for seat_ast in data:
    row_ast = seat_ast[:-3]
    col_ast = seat_ast[-3:]
    
    rows = [i for i in range(0,128)]
    cols = [i for i in range(0,8)]
    
    for char in row_ast:
        length = len(rows)
        pop_amount = int(length/2)
        counter = 0
        if char == 'B':
            while counter < pop_amount:
                throw_away = rows.pop(0)
                counter += 1
        if char == 'F':
            while counter < pop_amount:
                throw_away = rows.pop()
                counter += 1
    for x in rows:
        row_numbers.append(x)
    
    for char in col_ast:
        length = len(cols)
        pop_amount = int(length/2)
        counter = 0
        if char == 'L':
            while counter < pop_amount:
                throw_away = cols.pop()
                counter += 1
        if char == 'R':
            while counter < pop_amount:
                throw_away = cols.pop(0)
                counter += 1
    for x in cols:
        col_numbers.append(x)
        
for counter in range(len(row_numbers)):
    seat_id = (row_numbers[counter] * 8) + col_numbers[counter]
    seat_ids.append(seat_id)
```

```python
seat_ids.sort()
print(seat_ids)
```
